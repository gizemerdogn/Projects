```python
import pandas as pd
df = pd.read_csv("/Users/gizemerdogan/Downloads/exhibitA-input.csv", sep='\t')
df.head()
```

```python
# split PLAY_TS to use DATE column while filtering

df[["DATE", "TIME"]] = df["PLAY_TS"].str.split(' ', expand=True)
df = df.drop("PLAY_TS", axis=1)
df.head()
```

```python
# filter date column and find unique SONG_IDs per client

new_df = df[df["DATE"] == "10/08/2016"].groupby(["CLIENT_ID"], as_index=False)["SONG_ID"].value_counts()
new_df.head()
```

```python
# find distinct play counts per client

new_dict = {"CLIENT_ID" : new_df.CLIENT_ID.value_counts().index,
            "DISTINCT_PLAY_COUNT": new_df.CLIENT_ID.value_counts()}

new = pd.DataFrame(new_dict).reset_index(drop=True)
new.head()
```

```python
# find client count with respect to distinct play counts

new_dict_2 = {"DISTINCT_PLAY_COUNT" : new.DISTINCT_PLAY_COUNT.value_counts().index,
              "CLIENT_COUNT": new.DISTINCT_PLAY_COUNT.value_counts()}

new_2 = pd.DataFrame(new_dict_2).reset_index(drop=True)
new_2.head()
```
